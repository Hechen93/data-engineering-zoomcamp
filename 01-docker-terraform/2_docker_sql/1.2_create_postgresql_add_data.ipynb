```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

```python
import sqlalchemy as sa
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('yellow_tripdata_2021-01.csv', low_memory=False)
```

```python
df
```

```python
pd.io.sql.get_schema(df, name='yellow_taxi_data')
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data'))
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
engine = sa.create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
create_table_statement = pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine)
print(create_table_statement)
```

```python
df_iter = pd.read_csv('yellow_tripdata_2021-01.csv', iterator=True, chunksize=1000000)
```

```python
df = next(df_iter)
```

```python
df
```

```python
df.to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df.head(n=0)
```

```python
connection = engine.connect()

try:
    connection.execute(create_table_statement)
    print("Table created successfully.")
except Exception as e:
    print(f"Error creating table: {e}")
finally:
    connection.close()
```
